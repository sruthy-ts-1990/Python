# **1. Introduction**

Electricity is one of the most fundamental and accurate markers of a nation's development. It serves as a universal baseline for economic output, living standards, and technological advancement, which is why universal access is a primary focus of global initiatives like Sustainable Development Goal 7.

Electricity generation is a leading contributor to global greenhouse gases, accounting for roughly a quarter of all emissions. It is measured by carbon intensity (grams of CO₂ per kilowatt-hour), which varies significantly depending on whether a region relies on fossil fuels (like coal and gas) or zero-carbon sources (like nuclear, solar, and wind).

# **2. Objective**

Energy consumption and CO₂ emissions are inextricably linked, and burning fossil fuels (coal, oil, natural gas) to generate power remains the primary driver of global greenhouse gas (GHG) emissions. Reducing this reliance through renewable alternatives is essential for lowering our carbon footprint.


The objective of this project is to identify clean renewable  energy sources and their evolution from year 2000 to 2025 which is required to avoid the worst impacts of climate change.

# **6. Data Analysis**



**Formatting Data**

```python
#filtering out world CO2 Emission
df_Emissions = df_Yearly[(df_Yearly['Category']=='Power sector emissions') & (df_Yearly['Area']=='World') & (df_Yearly['Subcategory']=='Fuel')]
print(df_Emissions)
df_Emissions_wide = df_Emissions.pivot(index="Year", columns="Variable", values="Value")
df_Emissions_wide_world=df_Emissions_wide.add_suffix('_emission')
print(df_Emissions_wide_world.head())

#world CO2 Emission by Year
df_Emissions_world=df_Emissions.groupby('Year')['Value'].sum().reset_index().rename(columns={'Value': 'Total Emissions'})
print(df_Emissions_world)
```

```python
#filtering out world Electicity Generattion
df_yearly_Genaeration_World = df_Yearly[(df_Yearly['Category']=='Electricity generation') & (df_Yearly['Area']=='World') & (df_Yearly['Subcategory']=='Fuel') & (df_Yearly['Unit']!= '%')]
df_yearly_Genaeration_World_long = df_yearly_Genaeration_World[['Year','Variable','Unit','Value']]
print(df_yearly_Genaeration_World_long)

# Convert to wide format
df_yearly_Genaeration_World_wide = df_yearly_Genaeration_World_long.pivot(index="Year", columns="Variable", values="Value")
print(df_yearly_Genaeration_World_wide)

df_Generation_wide_world=df_yearly_Genaeration_World_wide.add_suffix('_generation')
print(df_Generation_wide_world.head())
```

```python
#Joining Emission table with Generation table based on column: year
df_merged=df_Generation_wide_world.join(df_Emissions_wide_world,on='Year', how='inner')
df_merged
```

```python
#unique fuels
print(df_yearly_Genaeration_World_long.head())
df_fuels=df_yearly_Genaeration_World_long['Variable'].drop_duplicates(ignore_index=True)
print(df_fuels)

#Emission intensity (or carbon intensity) is the rate of greenhouse gas (GHG) emissions released relative to a specific unit of activity, output, or economic value.
#Emission intensity= Emissions/Electricity Generation
#calculate Emission/kWh
  #1mT=10^7 g, 1 TWh= 10^9 kWh
for x in df_fuels:
  df_merged[x+'_Emission_per_kWh'] = (df_merged[x+'_emission'] / df_merged[x+'_generation'])*100
df_merged
```

```python
#long data format for Emission Vs Generation plot
df_Emissions_gen_merged = df_Yearly[((df_Yearly['Category']=='Power sector emissions')|(df_Yearly['Category']=='Electricity generation') )& (df_Yearly['Area']=='World') & (df_Yearly['Subcategory']=='Fuel')]
print(df_Emissions_gen_merged.head())
df_Emissions_gen_merged_sum=pd.DataFrame(df_Emissions_gen_merged.groupby(['Category','Subcategory','Year'])['Value'].sum())
print(df_Emissions_gen_merged_sum)
```

```python
# Top 10 countries CO2 Emission
df_Emissions_by_Country = df_Yearly[(df_Yearly['Category']=='Power sector emissions') & (df_Yearly['ISO 3 code']!=0) & (df_Yearly['Subcategory']=='Fuel')]
print(df_Emissions_by_Country)
df_Emissions_by_Country_sum=pd.DataFrame(df_Emissions_by_Country.query('Year==2025').groupby(['Area'])['Value'].sum().sort_values(ascending=False).head(10))
print(df_Emissions_by_Country_sum.index)
print(df_Emissions_by_Country_sum.columns)
```

```python
#calculating Average Emissions for each Fuel Type
sr_CO2_per_kWh =df_merged.loc[:, df_merged.columns.str.endswith('_kWh')].mean()#axis=1)

df_CO2_per_kWh_dump =pd.DataFrame(sr_CO2_per_kWh)
df_CO2_per_kWh = df_CO2_per_kWh_dump.reset_index(level='Variable')
res = list(df_CO2_per_kWh.columns)
print(res)

df_CO2_per_kWh['Fuel'] = df_CO2_per_kWh['Variable'].str.split('_').str[0]

df_CO2_per_kWh = df_CO2_per_kWh.rename(columns={0: "Value"})
df_CO2_per_kWh
```

```python
df_yearly_Genaeration_World_wide
```

```python
#Renewable energy generation

df_yearly_Renewable_Gen_wide = df_yearly_Genaeration_World_wide[['Hydro', 'Solar', 'Wind']]


df_yearly_Renewable_Gen_wide.reset_index(inplace=True)

df_yearly_Renewable_Gen_wide
df_yearly_Renewable_Gen_long = df_yearly_Renewable_Gen_wide.melt(
    id_vars=['Year'],
    value_vars=['Hydro', 'Solar', 'Wind'],
    var_name='Fuel',
    value_name='Value'
    ).sort_values(by='Year')
df_yearly_Renewable_Gen_long

```

# **8 Interpretation**

The global energy transition is the essential shift from fossil fuels to renewable energy sources to mitigate climate change and reduce carbon footprint and the ultimate aim of this transition is to achieve net-zero emissions.

Wind and solar are considered intermittent electricity sources, meaning they only produce electricity when the resource (wind or sun) is available, unlike dispatchable power sources-such as natural gas, coal, and nuclear power,which can operate as base-load generating units. Yet in 2025 Electricity from wind power and utility-scale solar power has increased to 17% and renewable energy sources combined provide 33.8% of total energy generated around the world.

 We are tackling a global need to reduce global warming, to reduce carbon and all greenhouse emissions, and have more renewable, stable, and flexible solutions that reflect the needs and demands of the 21st century, where we are highly dependent on a steady, sustainable source of energy supply.

  This analysis reveals the exponential growth of renewable energy sources which is major step to achieve net-zero emissions and the future looks bright.
